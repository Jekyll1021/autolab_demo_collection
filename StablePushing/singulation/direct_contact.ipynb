```python
%matplotlib inline

import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
first_contact_dist_lst = []
first_contact_pos_lst = []

for filename in os.listdir("primary_contact"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("primary_contact/"+filename, 'r') as f:
        dic = json.load(f)
    
    first_contact = dic['first contact object']
    first_contact_dist_lst.append(dic[str(first_contact)+" dist to pushing line"])
    first_contact_pos_lst.append(dic[str(first_contact)+" change of pos"])
    
```

```python
plt.scatter(first_contact_dist_lst, first_contact_pos_lst)
plt.plot(np.unique(first_contact_dist_lst), np.poly1d(np.polyfit(first_contact_dist_lst, first_contact_pos_lst, 1))(np.unique(first_contact_dist_lst)), 'r')
```

```python
com_dist_bins = [[], [], [], [], [], []]
change_pos_lst = [[], [], [], [], [], []]

for j in range(len(first_contact_dist_lst)):
    bin_num = int((first_contact_dist_lst[j]) // 0.2)
    com_dist_bins[bin_num].append(first_contact_dist_lst[j])
    change_pos_lst[bin_num].append(first_contact_pos_lst[j])
```

```python
for i in range(len(change_pos_lst)):
    weights = np.ones_like(change_pos_lst[i])/float(len(change_pos_lst[i]))
    plt.hist(change_pos_lst[i], weights=weights, bins=np.arange(0.0, 7, 0.2), alpha=0.4, label="com dist to pushing line = "+str(round(i * 0.2, 1)))
    plt.legend(loc='upper right')
```

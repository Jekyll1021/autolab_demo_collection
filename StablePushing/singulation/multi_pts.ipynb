```python
%matplotlib inline

import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
distribution = []
pos_change = []
distribution_multi_pts = []
pos_change_multi_pts = []
distribution_multi_pts_stable = []
pos_change_multi_pts_stable = []

for filename in os.listdir("comparison_multi_pts/distribution"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_multi_pts/distribution/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution.append(dic['mean separation after push'] - dic['mean separation before push'])

for filename in os.listdir("comparison_multi_pts/distribution_multi_pts"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_multi_pts/distribution_multi_pts/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change_multi_pts.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution_multi_pts.append(dic['mean separation after push'] - dic['mean separation before push'])
        
for filename in os.listdir("comparison_multi_pts/distribution_multi_pts_stable"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_multi_pts/distribution_multi_pts_stable/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change_multi_pts_stable.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution_multi_pts_stable.append(dic['mean separation after push'] - dic['mean separation before push'])
```

```python
weights = np.ones_like(distribution)/float(len(distribution))
plt.hist(distribution, weights=weights, bins=np.arange(-1, 4.3, 0.1), alpha=0.4, label="single point push")
weights2 = np.ones_like(distribution_multi_pts)/float(len(distribution_multi_pts))
plt.hist(distribution_multi_pts, weights=weights2, bins=np.arange(-1, 4.3, 0.1), alpha=0.4, label="two points push")
weights3 = np.ones_like(distribution_multi_pts_stable)/float(len(distribution_multi_pts_stable))
plt.hist(distribution_multi_pts_stable, weights=weights3, bins=np.arange(-1, 4.3, 0.1), alpha=0.4, label="two points push pointing COM")
plt.legend(loc='upper left')
```

```python
weights = np.ones_like(pos_change)/float(len(pos_change))
plt.hist(pos_change, weights=weights, bins=np.arange(-0.2, 7.3, 0.1), alpha=0.4, label="single point push")
weights2 = np.ones_like(pos_change_multi_pts)/float(len(pos_change_multi_pts))
plt.hist(pos_change_multi_pts, weights=weights2, bins=np.arange(-0.2, 7.3, 0.1), alpha=0.4, label="two points push")
weights3 = np.ones_like(pos_change_multi_pts_stable)/float(len(pos_change_multi_pts_stable))
plt.hist(pos_change_multi_pts_stable, weights=weights3, bins=np.arange(-0.2, 7.3, 0.1), alpha=0.4, label="two points push pointing COM")
plt.legend(loc='upper left')
```

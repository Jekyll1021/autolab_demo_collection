```python
%matplotlib inline

import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
overall = {}

for obj_num in os.listdir("prune/"):
    for group in os.listdir("prune/"+obj_num):
        for filename in os.listdir("prune/"+obj_num+"/"+group):
            prune_policy = filename.split(".")[0][2:]
            if group not in overall.keys():
                overall[group] = {}
            if prune_policy not in overall[group].keys():
                overall[group][prune_policy] = {}
            if obj_num not in overall[group][prune_policy].keys():
                overall[group][prune_policy][obj_num] = []
            with open("prune/" + obj_num+"/"+group+"/"+filename, 'r') as f:
                dic = json.load(f)
            overall[group][prune_policy][obj_num].append(dic['mean separation after push'] - dic['mean separation before push'])
```

```python
prune_0 = {}
for prune_policy in overall["0"]:
    if prune_policy not in prune_0.keys():
        prune_0[prune_policy] = {"mean":[], "std":[]}
    for i in range(3, 8):
        prune_0[prune_policy]["mean"].append(np.mean(overall["0"][prune_policy][str(i)]))
        prune_0[prune_policy]["std"].append(np.std(overall["0"][prune_policy][str(i)]))
```

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm']
plt.figure()
i = 0
for prune_policy in prune_0.keys():
    plt.errorbar(range(3, 8), prune_0[prune_policy]["mean"], yerr=prune_0[prune_policy]["std"], color=colors[i], label=prune_policy)
#     plt.plot(range(len(prune_0[prune_policy]["mean"])), prune_0[prune_policy]["mean"], color=colors[i], label=prune_policy)
    i += 1
plt.legend()
```

```python
colors = ['r', 'g', 'b', 'y', 'c', 'm']
base = np.array(prune_0["no_prunesummary"]["mean"])
base_std = np.array(prune_0["no_prunesummary"]["std"])
plt.figure()
i = 0
for prune_policy in prune_0.keys():
    plt.plot(range(3, 8), np.array(prune_0[prune_policy]["mean"])/base, color=colors[i], label=prune_policy)
#     plt.plot(range(len(prune_0[prune_policy]["mean"])), prune_0[prune_policy]["mean"], color=colors[i], label=prune_policy)
    i += 1
plt.legend()
```

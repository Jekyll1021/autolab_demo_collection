```python
%matplotlib inline

import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
distribution = []
pos_change = []
distribution_multi_pts = []
pos_change_multi_pts = []
distribution_multi_pts_stable = []
pos_change_multi_pts_stable = []
distribution_area_contact = []
pos_change_area_contact = []
distribution_area_contact_stable = []
pos_change_area_contact_stable = []

for filename in os.listdir("comparison_mode_distribution/distribution"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_mode_distribution/distribution/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution.append(dic['mean separation after push'] - dic['mean separation before push'])

for filename in os.listdir("comparison_mode_distribution/distribution_multi_pts"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_mode_distribution/distribution_multi_pts/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change_multi_pts.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution_multi_pts.append(dic['mean separation after push'] - dic['mean separation before push'])
        
for filename in os.listdir("comparison_mode_distribution/distribution_multi_pts_stable"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_mode_distribution/distribution_multi_pts_stable/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change_multi_pts_stable.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution_multi_pts_stable.append(dic['mean separation after push'] - dic['mean separation before push'])

for filename in os.listdir("comparison_mode_distribution/distribution_area"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_mode_distribution/distribution_area/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change_area_contact.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution_area_contact.append(dic['mean separation after push'] - dic['mean separation before push'])

for filename in os.listdir("comparison_mode_distribution/distribution_area_stable"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_mode_distribution/distribution_area_stable/"+filename, 'r') as f:
        dic = json.load(f)
    if dic['first contact object'] != -1:
        pos_change_area_contact_stable.append(dic[str(dic['first contact object'])+" change of pos"])
        distribution_area_contact_stable.append(dic['mean separation after push'] - dic['mean separation before push'])
```

```python
weights = np.ones_like(distribution)/float(len(distribution))
plt.hist(distribution, weights=weights, bins=np.arange(-1, 5, 0.1), alpha=0.4, label="single point push")
weights2 = np.ones_like(distribution_multi_pts)/float(len(distribution_multi_pts))
plt.hist(distribution_multi_pts, weights=weights2, bins=np.arange(-1, 5, 0.1), alpha=0.4, label="two points push")
weights3 = np.ones_like(distribution_multi_pts_stable)/float(len(distribution_multi_pts_stable))
plt.hist(distribution_multi_pts_stable, weights=weights3, bins=np.arange(-1, 5, 0.1), alpha=0.4, label="two points push pointing COM")
weights4 = np.ones_like(distribution_area_contact)/float(len(distribution_area_contact))
plt.hist(distribution_area_contact, weights=weights4, bins=np.arange(-1, 5, 0.1), alpha=0.4, label="area push")
weights5 = np.ones_like(distribution_area_contact_stable)/float(len(distribution_area_contact_stable))
plt.hist(distribution_area_contact_stable, weights=weights5, bins=np.arange(-1, 5, 0.1), alpha=0.4, label="area push pointing COM")
plt.legend(loc='upper left')
```

```python
change_pos_sum = []
delta = []
for filename in os.listdir("comparison_mode_distribution/distribution_multi_pts"):
    if len(filename.split("."))!=2 or filename.split(".")[1]!="json":
        continue
    with open("comparison_mode_distribution/distribution_multi_pts/"+filename, 'r') as f:
        dic = json.load(f)
    obj_set = set()
    obj_set.add(0)
    obj_set.add(1)
    obj_set.add(2)
    
    first_contact = dic['first contact object']
    obj_set.remove(first_contact)
    obj_lst = list(obj_set)
    
    s = 0
    
    for obj in obj_lst:
        s += (dic[str(obj)+" change of pos"])
    change_pos_sum.append(s)
    
    delta.append(dic['mean separation after push'] - dic['mean separation before push'])
```

```python
plt.scatter(change_pos_sum, delta)
plt.plot(np.unique(change_pos_sum), np.poly1d(np.polyfit(change_pos_sum, delta, 1))(np.unique(change_pos_sum)), 'r')
```
